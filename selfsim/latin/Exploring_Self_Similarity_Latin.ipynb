### Exploring Self-Similarity metric for Semantic Change detection (Latin)

```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

### Self-sim by layer comparison

```python
selfsim_1 = pd.read_csv('self_sim_by_layer_ISE1.csv', delimiter=',') # 2-1 century BC 
```

```python
selfsim_2 = pd.read_csv('self_sim_by_layer_ISE2.csv', delimiter=',') # 1-21 century AD
```

```python
selfsim_1.loc[selfsim_1['word'] == 'beatus']
```

```python
selfsim_2.loc[selfsim_2['word'] == 'beatus']
```

```python
def compare_plots(word_title):
    # Dataset
    x = np.array([i for i in range(1, 13)])
    y1 = np.array(selfsim_1.loc[selfsim_1['word'] == word_title].drop(['word', 'instances'], axis=1).values[0])
    y2 = np.array(selfsim_2.loc[selfsim_2['word'] == word_title].drop(['word', 'instances'], axis=1).values[0])

 
    cubic_model1 = interp1d(x, y1, kind = "cubic")
    cubic_model2 = interp1d(x, y2, kind = "cubic")

    X_ = np.linspace(x.min(), x.max(), 500)
    Y1_= cubic_model1(X_)
    Y2_= cubic_model2(X_)


    plt.plot(X_, Y1_)
    plt.plot(X_, Y2_, 'r')
    plt.title(f"Self-sim for {word_title}. Blue — old, red — new")
    plt.xlabel("layer")
    plt.ylabel("self-similarity")
    plt.show()
    
```

```python
compare_plots('acerbus') # change graded gold: 0.1693673835, binary:0
```
